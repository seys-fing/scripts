Las sinusoides discretas no necesariamente son periódicas, por ejemplo con frecuencia discreta $\theta_0=\sqrt{2}/2$ (en radianes)

 $$x[n]=\sin(\theta_0 n)=\sin(\sqrt{2}/2n)$$




```
  from scipy import signal
import numpy as np
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')




n=np.linspace(-5,20,26)


coseno=np.sin(np.sqrt(2)*n/4)

plt.stem(n,coseno)
plt.title(r'$\sin(\sqrt{2}/4 n)$')
plt.xlabel(r'$n$')

plt.show
  
```

Para que sea periódica  $$e^{j\theta_0 n}=e^{j\theta_0 (n+N)}$$ entonces $$e^{j\theta_0 N}=1$$  luego 

$$\theta_0 N=2\pi k \ \Rightarrow \ \theta_0=2\pi k/N$$

Por ejemplo podemos poner $\theta_0 = m*\pi/8$

Para analizarlo conviene usar exponenciales

$$e^{j (2\pi k+\theta_0) n}=e^{j 2\pi k n} e^{j \theta_0 n}=e^{j \theta_0 n}$$

Concluimos que frecuencias separadas por múltiplos de $2\pi$ dan la misma señal.

Por ejemplo 


$$e^{j 15 \pi  n/8} = e^{- j \pi n/8}= (e^{j \pi n/8})^*$$

entonces tienen la misma parte real



$$\cos(15 \pi  n/8) =\cos( \pi n/8).$$
