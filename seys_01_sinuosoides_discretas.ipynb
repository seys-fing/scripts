Las sinusoides discretas no necesariamente son peri√≥dicas, por ejemplo con frecuencia discreta $\theta_0=\sqrt{2}/2$ (en radianes)

 $$x[n]=\sin(\theta_0 n)=\sin(\sqrt{2}/2n)$$




```python
from scipy import signal
import numpy as np
import matplotlib.pyplot as plt
import warnings

warnings.filterwarnings('ignore')

n = np.linspace(-5, 20, 26)

coseno = np.sin(np.sqrt(2) * n / 4)

plt.stem(n, coseno)
plt.title(r'$\sin(\sqrt{2}/4 n)$')
plt.xlabel(r'$n$')

plt.show;
```

Para que sea peri√≥dica  $$e^{j\theta_0 n}=e^{j\theta_0 (n+N)}$$ entonces $$e^{j\theta_0 N}=1$$  luego 

$$\theta_0 N=2\pi k \ \Rightarrow \ \theta_0=2\pi k/N$$

Por ejemplo podemos poner $\theta_0 = m*\pi/8$


#Frecuencias 
¬øQu√© sucede cuando aumentamos la frecuencia?

```python
plt.figure()
coseno = np.cos(0 * n)
plt.stem(n, coseno)
plt.title(r'$cos(0 n)$')
plt.xlabel(r'$n$')
plt.show;

plt.figure()
coseno = np.cos(np.pi / 8 * n)
plt.stem(n, coseno)
plt.title(r'$cos(\pi\ n/8)$')
plt.xlabel(r'$n$')
plt.show;

plt.figure()
coseno = np.cos(2 * np.pi / 8 * n)
plt.stem(n, coseno)
plt.title(r'$cos(2*\pi\ n/8)$')
plt.xlabel(r'$n$')
plt.show;

plt.figure()
coseno = np.cos(4 * np.pi / 8 * n)
plt.stem(n, coseno)
plt.title(r'$cos(4*\pi\ n/8)$')
plt.xlabel(r'$n$')
freq = 1 / 4
tri = 2 * np.abs(signal.sawtooth(2 * np.pi * freq * n)) - 1
plt.plot(n, tri, '--r')
plt.show;

plt.figure()
coseno = np.cos(8 * np.pi / 8 * n)
plt.stem(n, coseno)
plt.title(r'$cos(8*\pi\ n/8)$')
plt.xlabel(r'$n$')
plt.show;
```

El $\cos(\pi n )$ es el coseno de mayor frecuencia posible en tiempo discreto. ¬øQu√© sucede si cotinuamos subiendo $m$?

```python
plt.figure()
coseno = np.cos(10 * np.pi / 8 * n)
plt.stem(n, coseno)
plt.title(r'$cos(10\pi n/8)$')
plt.xlabel(r'$n$')
plt.show;

plt.figure()
coseno = np.cos(12 * np.pi / 8 * n)
plt.stem(n, coseno)
plt.title(r'$cos(12\pi n/8)$')
plt.xlabel(r'$n$')
plt.show;

plt.figure()
coseno = np.cos(14 * np.pi / 8 * n)
plt.stem(n, coseno)
plt.title(r'$14 cos(\pi n/8)$')
plt.xlabel(r'$n$')
plt.show;

plt.figure()
coseno = np.cos(15 * np.pi / 8 * n)
plt.stem(n, coseno)
plt.title(r'$cos(15 \pi n/8)$')
plt.xlabel(r'$n$')
plt.show;

plt.figure()
coseno = np.cos(16 * np.pi / 8 * n)
plt.stem(n, coseno)
plt.title(r'$cos(16 \pi n/8)$')
plt.xlabel(r'$n$')
plt.show;
```

Para analizarlo conviene usar exponenciales $$e^{j (2\pi k+\theta_0) n}=e^{j 2\pi k n} e^{j \theta_0 n}=e^{j \theta_0 n}.$$
Concluimos que frecuencias separadas por m√∫ltiplos de $2\pi$ dan la misma se√±al.

Por ejemplo, $$e^{j 15 \pi  n/8} = e^{- j \pi n/8}= (e^{j \pi n/8})^*$$ entonces tienen la misma parte real
$$\cos(15 \pi  n/8) =\cos( \pi n/8).$$
